```python
class Number:
    def __init__(self,val):
        self.val = val

one = Number(1)
print(one.val)
```

```python
from dataclasses import dataclass
# Python me ek decorator hota hai jo classes ko simple aur clean banata hai.

# Iska main use hota hai jab aapko sirf data store karna ho.

@dataclass
class Numbers:
    value:int

two = Number(2)
print(two.value)
```
